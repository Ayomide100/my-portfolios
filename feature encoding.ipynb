### Convert age into age group

```python
# Use return inside a function as a substitute to print
def ageConverter(age):
    if(age < 18):
        return "Adolescent"
    elif(age <= 50):
        return "Adult"
    else:
        return "Old"
```

```python
# ageConverter(44)
```

```python
# Column values get converted into the function you pass into the apply method for the column
df["age"].apply(ageConverter)
```

```python
df["ageGroup"]  = df["age"].apply(ageConverter)
```

```python
df.head()
```

```python
def bmiGroup(bmi_value):
    if(bmi_value <= 18.5):
        return "Underweight"
    elif(bmi_value <= 24.9):
        return "Healthy"
    elif(bmi_value <= 29.9):
        return "Overweight"
    else:
        return "Obese"
```

```python
df['bmi_range'] =  df.bmi.apply(bmiGroup)
```
