### Face detection function

```python
def face_detection(image):
    #read the image
    img = cv2.imread(image)
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    face_cascade = cv2.CascadeClassifier("haarcascade_frontalface_default.xml")
    faces = face_cascade.detectMultiScale(gray, 1.3, 5)
    x,y,w,h = faces[0]
    face_img = cv2.rectangle(img, (x, y), (x+w, y+h), (0, 255, 0), thickness = 2)
    return plt.imshow(face_img)
```

```python
import cv2
import os
import numpy as np
import faceRecognition as fr
```

```python
# Load images and label for traning data
faces, face_label = fr.labels_for_training_data("Images Dataset")
```

```python
#Show converted images label
face_label
```

```python
# Train the classifier on the faces and face_label
FR = fr.train_classifier(faces, face_label)
```

```python
names = {0: "Lionel messi", 1: "Maria Shirapova", 2: "Roger Federer", 3: "Serena williams", 4: "Virat kohli"}
```

```python
cap = cv2.VideoCapture(0)
while True:
    ret, frame = cap.read()
    if not ret:
        break
    faces_detected, gray_img = fr.faceDetection(frame)
    for face in faces_detected:
            (x,y,w,h)=face
            roi_gray=gray_img[y:y+h,x:x+h]
            label,confidence=FR.predict(roi_gray)#predicting the label of given image
            print("confidence:",confidence)
            print("label:",label)
            fr.draw_rect(frame,face)
            # cv2.rectangle(frame, (x, y), (x+w, y+h), (0, 255, 0), thickness=2)
            if confidence < 50:
                predicted_name = names[label]  # Retrieve the name from the dictionary using the predicted label
                fr.put_text(frame, predicted_name,x,y)
            else:
                fr.put_text(frame,"Not recognized",x,y)
        
    resized_img=cv2.resize(frame,(1000,1000))
    cv2.imshow("frame", resized_img)
    #Wait for key event for 1 milliseconds
    if cv2.waitKey(50) & 0xFF == ord('q'):
        break

#Release video capture object and close all windows
cap.release()
cv2.destroyAllWindows()
```
