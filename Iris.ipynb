```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("IRIS.csv")
```

```python
df.head()
```

```python
df["species"]. unique()
```

```python
df.isnull().sum()
```

```python
df["species"]= df["species"].map({'Iris-setosa' : 1, 'Iris-versicolor': 2,'Iris-virginica' : 3})
```

```python
df.head()
```

```python
X = df.drop("species", axis = 1)
Y = df["species"]
```

```python
X.shape
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X = scaler.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import f1_score
```

```python
train_x, test_x, train_y, test_y = train_test_split(X, Y, random_state= 12, test_size= 0.35)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
# Initialize the model
DT = DecisionTreeClassifier()
DT.fit(train_x, train_y)
```

```python
print(DT.score(train_x, train_y))
print(DT.score(test_x, test_y))
```

```python
predicted_values = DT.predict(test_x)
```

```python
from sklearn.metrics import classification_report
```

```python
report = classification_report(test_y, predicted_values)
```

```python
print(report)
```

```python
f1_score(predicted_values, test_y, average= "macro")
```
